```python
import pandas as pd
import numpy as np
```

```python
%%bash
wget -cq https://archive.ics.uci.edu/ml/machine-learning-databases/00312/dow_jones_index.zip
unzip -o dow_jones_index.zip
# head -5 dow_jones_index.data
```

```python
%%bash
head -5 dow_jones_index.data
```

```python
df = pd.read_csv(
    "dow_jones_index.data",
    sep=",",
    header=0,
    
    #parsering
    decimal = ".",
    
    converters = {
        'open': lambda x: float(x.strip("$")),
        'high': lambda x:float(x.strip("$")),
        'low': lambda x:float(x.strip("$")),
        'close': lambda x:float(x.strip("$"))
    },
    
    # date time
    parse_dates = ['date'],
    date_parser= lambda x: pd.to_datetime(x)
)
```
